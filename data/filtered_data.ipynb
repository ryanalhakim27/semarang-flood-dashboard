```python
import pandas as pd
from shapely.geometry import Point, Polygon

# Load your rainfall data
rainfall_data = pd.read_csv('rainfall_data.csv')

# Define your boundary polygon (lon, lat order!)
boundary_coords = [
    (110.26058, -7.16924),
    (110.52927, -7.16689),
    (110.53222, -6.92284),
    (110.26007, -6.92160)
]
boundary_polygon = Polygon(boundary_coords)

# Function to check if a point is inside the polygon
def is_inside(lat, lon):
    return boundary_polygon.contains(Point(lon, lat))

# Apply the function to filter data
rainfall_data['inside'] = rainfall_data.apply(lambda row: is_inside(row['Lat_DD'], row['Lon_DD']), axis=1)

# Keep only rows inside the boundary
filtered_data = rainfall_data[rainfall_data['inside']].drop(columns=['inside'])

# Save filtered data
filtered_data.to_csv('rainfall_data_filtered.csv', index=False)

print(f"Original data: {len(rainfall_data)} rows")
print(f"Filtered data: {len(filtered_data)} rows")
```
